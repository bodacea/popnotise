# Explore early Twitter dataset

Basically, the dataset created by pulling Tweets in json, and converting those straight into a pandas dataframe

```python
import pandas as pd

df = pd.read_csv('qanon_tweets.csv')
df
```

```python
import json
ss = df['user'].iloc[0].replace("'", '"').replace(': None,', ': "",').replace(': False,', ': "False",').replace(': True,', ': "True",')
jj = json.loads(ss)
jj
```

```python
df['user'] = df['user'].str.replace("'", '"').str.replace(': None,', ': "",').str.replace(': False,', ': "False",').str.replace(': True,', ': "True",')
df['user'].apply(json.loads)                                                                                                                                    
```

```python
for i in range(len(df)):
    print('{}: {}'.format(i, json.loads(df['user'].iloc[i])))
```
