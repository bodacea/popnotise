```python
# Clean up the linefeed problem, e.g. user-generated text with linefeeds in breaks the
# CSV write.  Do this by adding a code where the user-generated linefeeds were, and 
# only outputting a carriage return for the start of each 'real' line of data.
infile = '../data/qanon_tweets/raw_scrapes/#qanon_tweets_2018_04_01_23_17_59_637556_980689639148310529_980673645273145343.csv'
fin = open(infile, 'r')
fout = open('outtest.csv', 'w')
headers = fin.readline()
fout.write(headers)
for line in fin:
    if line[:2] != ',,':
        print('Found badguy: {}\n'.format(line))
        #line = "%0A" + line #add url encoding for linefeed
    else:
        line = "\n" + line #add linefeed - this line is normal
    fout.write(line)
    
fin.close()
fout.close()
```
