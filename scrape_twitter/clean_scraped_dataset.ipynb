# Clean up scraped dataset

Because it would be a shame to pull all that data again

```python
import pandas as pd
import csv
import os
import re
import glob

pd.set_option("display.max_columns", 1000)

# Clean up the linefeed problem, e.g. user-generated text with linefeeds in breaks the
# CSV write.  Do this by adding a code where the user-generated linefeeds were, and 
# only outputting a carriage return for the start of each 'real' line of data.

indir = '../data/qanon_tweets/raw_scrapes/'
outdir = '../data/qanon_tweets/dataset/'
for infile in glob.glob(indir + '*.csv'):
    print(infile)
    fin = open(infile, 'r')
    outfile = outdir+infile[len(indir):-4]+'_clean.csv'
    fout = open(outfile, 'w')
    headers = fin.readline()
    fout.write(headers)
    for line in fin:
        line = re.sub(r'[\r\n]', '', line)
        if line[:2] == ',,':
            line = "\n" + line # this line is normal: add linefeed
        fout.write(line)

    fin.close()
    fout.close()
```

## Create cutdown dataset for data.world

```python
dfall = pd.DataFrame([])
for infile in glob.glob(outdir + '*.csv'):
    print(infile)
    df = pd.read_csv(infile, usecols=['id', 'user_name', 'user_id', 'created_at', 
                                      'text', 'retweet_count', 'source'])
    df['client'] = df['source'].str.extract('\"\>(.*)\<\/a\>', expand=False)
    df.drop(['source'], axis=1, inplace=True)
    dfall = pd.concat([dfall, df])
    dfall.to_csv('../data/qanon_tweets/qanon_all_tweets.csv', index=False, quoting=csv.QUOTE_ALL)
dfall
```
